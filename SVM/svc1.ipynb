```python
from sklearn.compose import ColumnTransformer
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split,GridSearchCV
df = pd.read_csv('dataset_traffic_accident_prediction1.csv')
```

```python
df.head()
```

```python
df = df.dropna(subset=['Accident'])
```

```python
X = df.drop('Accident',axis=1)
y = df['Accident']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
obj_cols = xtrain.select_dtypes(include='object').columns
num_cols = xtrain.select_dtypes(exclude='object').columns
```

```python
num_preprocessor = Pipeline(
    steps=[
        ('num_impute',SimpleImputer(strategy='mean')),
        ('scalar',StandardScaler())
    ]
)
cat_preprocessor = Pipeline(
    steps=[
        ('obj_impute',SimpleImputer(strategy='constant',fill_value="Unknown")),
        ('encode',OneHotEncoder(sparse_output=False))
    ]
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num_preprocessor',num_preprocessor,num_cols),
        ('cat_preprocessor',cat_preprocessor,obj_cols),
    ]
)
```

```python
pipeline=Pipeline(
    steps=[
        ('preprocessor',preprocessor),
        ('model',SVC())
    ]
)
```

```python
params = {
    'model__C':[0.01,0.1,1,10,100],
    'model__kernel':['rbf','sigmoid']
}
```

```python
g_cv = GridSearchCV(pipeline,params,cv=5,n_jobs=-1)
```
