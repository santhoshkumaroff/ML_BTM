```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score



```

```python
df=pd.read_csv('loan_approval_dataset.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
scalar=StandardScaler()
encoder=OneHotEncoder(sparse_output=False)
```

```python
numcol=df.select_dtypes(exclude='object').columns
catcol=df.select_dtypes(include='object').drop('Loan_Approved',axis=1).columns
x=df.drop('Loan_Approved',axis=1)
y=df['Loan_Approved']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
xtrain
```

```python
catcol
```

```python
encoder.fit_transform(xtrain[catcol])
```

```python
xtrain[numcol]=scalar.fit_transform(xtrain[numcol])
xtrain[encoder.get_feature_names_out()]=encoder.fit_transform(xtrain[catcol])
xtrain.drop(catcol,axis=1,inplace=True)
```

```python
model=LogisticRegression(class_weight='balanced')
xtrain
```

```python
model.fit(xtrain,ytrain)
```

```python
xtest[numcol]=scalar.transform(xtest[numcol])
xtest[encoder.get_feature_names_out()]=encoder.transform(xtest[catcol])
xtest.drop(catcol,axis=1,inplace=True)
```

```python
ytest.value_counts()
```

```python
ypred=model.predict(xtest)
```

```python
confusion_matrix(ytest,ypred)
```

```python
precision_score(ytest,ypred,pos_label='No')
```

```python
precision_score(ytest,ypred,pos_label='Yes')
```

```python
recall_score(ytest,ypred,pos_label='No')
```

```python
recall_score(ytest,ypred,pos_label='Yes')
```

```python
accuracy_score(ytest,ypred)
```
