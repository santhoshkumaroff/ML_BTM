```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler
from sklearn.metrics import confusion_matrix,recall_score,precision_score,f1_score,accuracy_score
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
df.head()
```

```python
X=df.drop('loan_status',axis=1)
y=df['loan_status']
```

```python
num_cols = X.select_dtypes(exclude="object").columns
cat_cols = X.select_dtypes(include="object").columns
```

```python
df[num_cols]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
scaler = StandardScaler()
encode = OneHotEncoder(sparse_output=False)
```

```python
xtrain
```

```python
xtrain[num_cols] = scaler.fit_transform(xtrain[num_cols])
```

```python
xtrain
```

```python
x_train_encoded_values = encode.fit_transform(xtrain[cat_cols])
x_train_encoded_values
```

```python
xtrain[encode.get_feature_names_out()] = x_train_encoded_values
```

```python
xtrain
```

```python
xtrain.drop(cat_cols,axis=1,inplace=True)
```

```python
xtrain
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
xtest
```

```python
xtest[num_cols] = scaler.transform(xtest[num_cols])
```

```python
xtest[encode.get_feature_names_out()] = encode.transform(xtest[cat_cols])
```

```python
xtest
```

```python
xtest.drop(cat_cols,axis=1,inplace=True)
```

```python
xtest
```

```python
model.score(xtest,ytest)
```

```python
ypred = model.predict(xtest)
```

```python
ytest.value_counts()
```

```python
confusion_matrix(ytest,ypred)
```

```python
precision_score(ytest,ypred,pos_label=1)
```

```python
precision_score(ytest,ypred,pos_label=0)
```

```python
recall_score(ytest,ypred,pos_label=1)
```

```python
recall_score(ytest,ypred,pos_label=0)
```

```python
f1_score(ytest,ypred,pos_label=1)
```

```python
f1_score(ytest,ypred,pos_label=0)
```
